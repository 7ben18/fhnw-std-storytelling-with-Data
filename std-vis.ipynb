```python
# import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import plotly.express as px
import plotly.graph_objects as go

```

```python
# read xlsx file with pandas

df_jahr = pd.read_excel("Benzin-Diesel-data.xlsx", sheet_name="Jahresansicht")
df_monat = pd.read_excel("Benzin-Diesel-data.xlsx", sheet_name="Monatsansicht")

# convert Jahr column into datetime
df_jahr["Jahr"] = pd.to_datetime(df_jahr["Jahr"], format="%Y")

# conert Monat column into datetime
df_monat["Monat"] = pd.to_datetime(df_monat["Monat"], format="%Y-%m-$d")

# head
display("jahr", df_jahr.head())
display("monat", df_monat.head())
```

```python
# create a function for plotting the first three column as variable name, parameter are: dataframe, three column name and title
def plot(df, zeitformat, col1, col2, col3, title):
    '''
    df: df_jahr oder df_monat
    zeitformat: Jahr oder Monat
    col1: Benzin 95
    col2: Benzin 98
    col3: Diesel
    Title: irgendaws

    '''
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=df[zeitformat], y=df[col1], name="Benzin 95"))
    fig.add_trace(go.Scatter(x=df[zeitformat], y=df[col2], name="Benzin 98"))
    fig.add_trace(go.Scatter(x=df[zeitformat], y=df[col3], name="Diesel"))
    fig.update_layout(title=title, xaxis_title="Jahr", yaxis_title="Preis in CHF", legend_title="Treibstoff")
    # add source annotation below the plot
    fig.add_annotation(text="Quelle: Bundesamt für Statistik", xref="paper", yref="paper", x=-0.005, y=-0.15, showarrow=False)
    
    fig.show()


```

```python
plot(df_monat, "Monat", "Bleifrei 95", "Bleifrei 98", "Diesel", "Trebstoffpreis in Franken zum Monatsende")

plot(df_monat, "Monat",  'Bleifrei 95 / LIK ', 'Bleifrei 98 / LIK', 'Diesel / LIK', "Treibstoffpreis in Franken zum Monatsende, Inflationsbereinigt")

plot(df_monat, "Monat", 'Bleifrei 95 / LIK - Basis 12/2015', 'Bleifrei 98 / LIK - Basis 12/2015', 'Diesel / LIK - Basis 12/2015', "Treibstoffpreis in Franken zum Monatsende, Inflationsbereinigt, Relativiert auf Basis 12/2015")
```
